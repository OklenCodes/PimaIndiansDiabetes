```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```python
ds = loadtxt('pima-indians-diabetes.csv', delimiter=',')
```

```python
X = ds[:,0:8] #First 8 columns
y = ds[:,8] #9th column
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, input_dim=8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) #Binary cross entropy as it is a classification problem
model.fit(X, y, epochs=150, batch_size=8)
```

Only 75% Accuracy On 150 Epochs and 10 Batches. Accuracy doesnt change much when figures are adjusted. Would need more Data

```python
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```
